```python
%matplotlib inline
from urllib2 import Request, urlopen
import json
from pandas.io.json import json_normalize
import pandas as pd
```

## Download metainformation about images (statistical maps)

```python
request=Request('http://neurovault.org/api/images/?format=json')
response = urlopen(request)
elevations = response.read()
data = json.loads(elevations)
images_df = json_normalize(data)
images_df['collection'] = images_df['collection'].apply(lambda x: int(x.split("/")[-2]))
images_df.rename(columns={'collection':'collection_id'}, inplace=True)
images_df['id'] = images_df['url'].apply(lambda x: int(x.split("/")[-2]))
images_df = images_df.set_index("id")
```

## Pick maps from stop signal tasks

```python
stop_signal_df = images_df[images_df.cognitive_paradigm_cogatlas.str.contains("stop signal") == True]
stop_signal_df
```

## Pick only maps contrasting "stop" and "go" conditions

```python
stop_signal_df.contrast_definition.unique()
```

```python
stop_vs_go_contrast_labels = [u'succstop minus go', 
                              u'successful stop - go', 
                              u'succ stop vs go', 
                              u'succcessful stop - go noncrit', 
                              u'successful stop- go noncrit']
stop_vs_go_contrast = stop_signal_df[stop_signal_df.contrast_definition.isin(stop_vs_go_contrast_labels)]
stop_vs_go_contrast
```

## Download and visualise the files

```python
import os, urllib
from nilearn.plotting import plot_glass_brain
analysis_dir = "stop_vs_go_metaanalysis"
if not os.path.exists(analysis_dir):
    os.makedirs(analysis_dir)
    
download_cache = os.path.join(analysis_dir, "cache")
if not os.path.exists(download_cache):
    os.makedirs(download_cache)

testfile = urllib.URLopener()
for row_idx, row in stop_vs_go_contrast.iterrows():
    print "Downloading image %d"%row_idx
    testfile.retrieve(row["file"], os.path.join(download_cache, "%04d.nii.gz"%row_idx))
    plot_glass_brain(os.path.join(download_cache, "%04d.nii.gz"%row_idx), title="Image %04d"%row_idx)
    
```

## Convert T maps to Z maps

```python
import scipy
scipy.stats.t.sf?
def t_to_z(t_scores, df):
    p_values = scipy.stats.t.sf(t_scores, df = df)
    z_values = scipy.stats.norm.isf(p_values)
    return z_values, p_values
```

## Average maps from the same study

```python
def get_number_of_subjects(collection_id):
    request=Request('http://neurovault.org/api/collections/%d?format=json'%collection_id)
    response = urlopen(request)
    elevations = response.read()
    data = json.loads(elevations)
    df = json_normalize(data)
    return int(df.number_of_subjects[0])
```

```python
get_number_of_subjects(42)
```

```python
from nilearn.image import mean_img
import nibabel as nb
study_maps_dir = os.path.join(analysis_dir, "study_maps")
if not os.path.exists(study_maps_dir):
    os.makedirs(study_maps_dir)

mean_maps = []
p_datas = []
z_datas = []
for collection_id in stop_vs_go_contrast.collection_id.unique():
    print "collection %d"%collection_id
    study_maps = stop_vs_go_contrast[stop_vs_go_contrast.collection_id == collection_id]
    image_files = [ os.path.join(download_cache, "%04d.nii.gz"%i) for i,row in study_maps.iterrows()]
    
    # convert t to z
    image_niis = [nb.load(f) for f in image_files]
    image_z_niis = []
    df = get_number_of_subjects(collection_id)-2
    print "Degrees of freedom = %d"%df
    for nii in image_niis:
        data_z, data_p = t_to_z(nii.get_data(), df = df)
        p_datas.append(data_p)
        z_datas.append(data_z)
        nii_z = nb.Nifti1Image(data_z, nii.get_affine())
        image_z_niis.append(nii)
    mean_map = mean_img(image_z_niis)
    plot_glass_brain(mean_map, title="Collection %04d mean map"%collection_id)
    mean_maps.append(mean_map)
    
```

## Fisher’s Z-score method (on 8 maps)

```python
import numpy as np
import math
cut_coords = [-15, -8, 6, 30, 46, 62]
z_meta8_data = np.array(z_datas).sum(axis=0)/math.sqrt(len(z_datas))
nii = nb.Nifti1Image(z_meta8_data, mean_map_nii.get_affine())
plot_stat_map(nii, display_mode = 'z', threshold=6, cut_coords=cut_coords, vmax=12)
```

```python
import numpy as np
import math
z_input_datas = [nii.get_data() for nii in mean_maps]
z_meta4_data = np.array(z_input_datas).sum(axis=0)/math.sqrt(len(z_input_datas))
nii = nb.Nifti1Image(z_meta4_data, mean_map_nii.get_affine())
p = plot_stat_map(nii, display_mode = 'z', threshold=6, cut_coords=cut_coords, vmax=12)
nii.to_filename("fisher_z_4_maps_stop_minus_go.nii.gz")
p.savefig("fisher_z_4_maps_stop_minus_go.pdf")
```
